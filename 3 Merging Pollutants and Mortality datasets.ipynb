```python
aggregated_deaths_by_gender = geo_df_clean.groupby(["State", "Gender", "Year"]).agg(
    {
        "Deaths": "sum",
        "Population": "sum",
        "Emissions": "first",
        "latitude": "first",
        "longitude": "first"
    })
aggregated_deaths_by_gender["Deaths_per_10,000"] = (aggregated_deaths_by_gender["Deaths"] / aggregated_deaths_by_gender["Population"]) * 10000

aggregated_deaths_by_gender.reset_index().to_csv("deaths_by_state_and_gender.csv", index=False)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
aggregated_deaths_by_gender
```

```python
# Model 1:
model_df = aggregated_deaths_by_gender[aggregated_deaths_by_gender["Emissions"].notna()]
model1 = LinearRegression()
X1 = model_df[["Emissions"]].copy()
y1 = model_df['Deaths_per_10,000']
model1.fit(X1, y1)
print('Model 1 Coefficients:', model1.coef_)
print('Model 1 Intercept:', model1.intercept_)
```

```python
# Model 2:
model_df2 = model_df.reset_index("Gender").copy()
model_df2["Gender"] = pd.get_dummies(model_df2['Gender'], drop_first=True)
model2 = LinearRegression()
X2 = model_df2[["Emissions", "Gender"]].copy()
y2 = model_df2['Deaths_per_10,000']
model2.fit(X2, y2)
print('Model 2 Coefficients:', model2.coef_)
print('Model 2 Intercept:', model2.intercept_)
```

```python
years = model_df2.reset_index("Year")["Year"]
years_train = (years < 2016).values
X2_train = X2[years_train]
y2_train = y2[years_train]
X2_test = X2[~years_train]
y2_test = y2[~years_train]
model2.fit(X2_train, y2_train)
print('Model 2 Coefficients:', model2.coef_)
print('Model 2 Intercept:', model2.intercept_)



# Predicting the outcome
predictions = model2.predict(X2_test)

# Evaluating accuracy
mse = mean_squared_error(y2_test, predictions)
print("Mean Squared Error:", mse)
```
